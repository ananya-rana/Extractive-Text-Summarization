```python
## preprocessing and counting word frequencies ##

word_freq ={}
for word in doc:
    if word.pos_ not in ["SPACE","X","PUNCT"] and word.text.lower() not in stopwords and word.like_url==False:
        if word.text not in word_freq.keys():
            word_freq[word.text]=1
        else:
            word_freq[word.text]+=1
```

```python
word_freq
```

```python
max_freq = max(word_freq.values())
max_freq
```

Although it seems to be a large number, but we have already removed stopwords and the length of the original text was in the range of 40,000 words. Therefore max_frequency is an acceptable number in this context as it will help us determine the relative importance of sentences.

```python
key_list = list(word_freq.keys())
val_list = list(word_freq.values())

position = val_list.index(max_freq)
print(key_list[position])
```

As you can see, for our example text "Neo" is a very important word, and whatever summary that is finally generated should have a higher probability than not to retain sentences which contain Neo.

```python
## calculating normalized word frequencies ##
for word in word_freq.keys():
    word_freq[word] = word_freq[word]/max_freq
```

```python
word_freq
```

Now for calculating the most imporatant sentences by adding normalized word frequencies in each sentence.

```python
sent_tokens = [sent for sent in doc.sents]
```

```python
sent_tokens
```

```python
len(sent_tokens)
```

```python
sent_scores = {}
for sent in sent_tokens:
    for word in sent:
        if word.text.lower() in word_freq.keys():
            if sent not in sent_scores.keys():
                sent_scores[sent] = word_freq[word.text.lower()]
            else:
                sent_scores[sent] +=word_freq[word.text.lower()]
```

The procedure followed below is a heuristic devised to retain sentences which achieve a specific required score, according to the requirement of the user.

```python
sumValues = 0
for sent in sent_scores:
    sumValues += sent_scores[sent]

sumValues
```

```python
# Average score of a sentence from the original text
   
avg_score = sumValues / len(sent_scores)
avg_score
```

```python
# Storing sentences into our summary.
# deg_of_sum here means degree/extent of summarization -> it can take a decimal value.

def summarize(deg_of_sum):
  summary = ''
  for sentence in sent_tokens:
    if (sentence in sent_scores) and (sent_scores[sentence] > (deg_of_sum * avg_score)):
        summary += " " + sentence.text
  print("LENGTH OF SUMMARY",len(summary))
  print(summary)
```

The problem with other solutions to this problem was that they were not able to retain the order of the statement, thus it resulted in summary being printed in a jumbled order. I have removed this specific issue and therefore the summary generated is ordered. I have given the users the option to decide the how much do they want to summarize. 
The parameters in the function "summarize" define the extent to which the user wants to summarize, if she wants to retain very important sentences she can set a high value and if she wants to retain more information, she can set it at lower value.

```python
summarize(2)
```

```python
## another summary which retains more information ##
summarize(1.3)
```

Finally, the problem with extractive text summarization is that it cannot be used to summarize topics which are more philosophical in nature since continuity that is required in the abstract process may get interrupted by skipping of a few sentences here and there.
